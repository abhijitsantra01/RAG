```python
### Text Loader
from langchain_community.document_loaders import TextLoader
from langchain_community.document_loaders import PyPDFLoader
```

```python
from langchain_community.document_loaders import PyPDFLoader

# Initialize the loader with the file path
pdf_loader = PyPDFLoader("../data/pdf_files/Soil Health Analysis and Crop Suitability Prediction Using ML.pdf")

# Load the documents (each page as a document)
pages = pdf_loader.load()

# Or load and split into chunks immediately
pages = pdf_loader.load_and_split() 

# You can now process the loaded pages, for example:
for page in pages:
    print(f"Page content snippet: {page.page_content[:100]}...")
    print(f"Metadata: {page.metadata}")
```
